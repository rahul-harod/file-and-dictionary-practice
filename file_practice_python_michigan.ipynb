# WORK with csv files

this is text file but has csv formate like equal no. of commas in each row. print only who won a medal

```python
olypmicsfile = open("F:\\olympicdata.txt", "r")
lines=olypmicsfile.readlines()
header=lines[0].strip().split(",")
print(header[0]+":",header[4]+";",header[5])
for line in lines[1:]:
    vals=line.strip().split(",")
    if vals[5] != "NA":
        print("{}: {}; {}".format(
                 vals[0],
                 vals[4],
                 vals[5]))
olypmicsfile.close()
```

```python
olympians = [("John Aalberg", 31, "Cross Country Skiing"),
             ("Minna Maarit Aalto", 30, "Sailing"),
             ("Win Valdemar Aaltonen", 54, "Art Competitions"),
             ("Wakako Abe", 18, "Cycling")]

outfile = open("F:\\reduced_olympics.csv", "w")
# output the header row
outfile.write('Name,Age,Sport')
outfile.write('\n')
# output each of the rows:
for olympian in olympians:
    row_string = '{},{},{}'.format(olympian[0], olympian[1], olympian[2])
    outfile.write(row_string)
    outfile.write('\n')
outfile.close()
```

Challenge: Using the file school_prompt.txt, if the character ‘p’ is in a word, then add the word to a list called p_words.

```python
with open("school_prompt.txt","r") as f:
    p_words=[]
    for line in f:
        sp=line.split()
        for word in sp:
            if "p" in word:
                p_words.append(word)
```

We have provided a file called emotion_words.txt that contains lines of words that describe emotions. Find the total number of words in the file and assign this value to the variable num_words.

```python
with open("emotion_words.txt","r") as f:
    lines=f.readlines()
    count=0
    for line in lines:
        words=line.split()
        l=len(words)
        count+=l
    num_words=count
```
